```python
import pandas as pd
```

```python
fin = pd.read_csv("../data/raw/aws_financials.csv")
```

```python
# Calculate revenue growth and ROI
fin["revenue_growth"] = fin["aws_revenue"].pct_change()
fin["roi"] = fin["aws_revenue"] / fin["capex"]
```

```python
# Categorize utilization rate
fin["util_category"] = fin["utilization_rate"].apply(
    lambda x: "High" if x > 75 else "Medium" if x > 50 else "Low"
)
```
